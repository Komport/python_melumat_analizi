**DataFrame**, potensial olaraq müxtəlif növ sütunları olan __2D__ yazılı məlumat quruluşudur. Bunu bir elektron cədvəl və ya SQL cədvəlinin bir hissəsi kimi düşünə bilərsiniz.
Init signature: `DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)`
>`data`: *özü numpy ndarray,dictionary,list,DataFrame,Seriyalar,sabitlər və.s özündə cəmləyə bilir və bir kod kimi işlədir*

>`index`: * çərçivə (gövdə) yıqılması üçün istifadə edilir əgər indeks yazılmasa adəti üzrə np.arange(n) olacaq*

>`columns`: * etiketləri çərçivə nəticəsində istifadə edilir əgər indeks yazılmasa adəti üzrə np.arange(n) olacaq*

```python
text3 = DataFrame(text2,columns=text2.columns[:4])
text3
```
